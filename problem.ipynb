```python
def fibonacci_series(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b
for i in fibonacci_series(10):
  print(i)
```

```python
def test_fib1():
  a,b=0,1
  while True:
    yield a
    a,b=b,a+b
```

```python
fib=test_fib1()
```

```python
def count_test(n):
  count=1
  while count<=n:
    yield count
    count=count+1
for i in count_test(10):
  print(i)
```
