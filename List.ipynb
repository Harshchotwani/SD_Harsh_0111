```python
# SRMCEM College Branch Selection and Fee Discount System

print("=" * 50)
print( Welcome to SRMCEM College")
print("=" * 50)

branches = {
    1: {
        "name": "Computer Science Engineering (CSE)",
        "fee": 120000,
        "discount": 20000
    },
    2: {
        "name": "Artificial Intelligence & Machine Learning (AI & ML)",
        "fee": 125000,
        "discount": 25000
    },
    3: {
        "name": "Information Technology (IT)",
        "fee": 115000,
        "discount": 15000
    },
    4: {
        "name": "Electronics & Communication Engineering (ECE)",
        "fee": 110000,
        "discount": 12000
    },
    5: {
        "name": "Mechanical Engineering (ME)",
        "fee": 100000,
        "discount": 10000
    },
    6: {
        "name": "Civil Engineering (CE)",
        "fee": 95000,
        "discount": 8000
    }
}

print("\nAvailable Branches:\n")

for key, value in branches.items():
    print(f"{key}. {value['name']}")

choice = int(input("\nEnter the branch number: "))

if choice in branches:
    branch = branches[choice]

    original_fee = branch["fee"]
    discount = branch["discount"]
    final_fee = original_fee - discount

    print("\n" + "=" * 50)
    print("Admission Details")
    print("=" * 50)
    print("Selected Branch :", branch["name"])
    print("Original Fee :", original_fee)
    print("Scholarship : ", discount)
    print("Final Fee : ", final_fee)
    print("=" * 50)

else:
    print("Invalid Branch Selection!")
```
