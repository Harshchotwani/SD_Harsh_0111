```python
# ATM Machine Program

card_number = "123"
pin = "1234"
balance = 50000

print("===================================")
print("      WELCOME TO XYZ BANK ATM      ")
print("===================================")

# Card Number
entered_card = input("Enter your Card Number: ")

if entered_card == card_number:

    # PIN Verification
    entered_pin = input("Enter your 4-digit PIN: ")

    if entered_pin == pin:

        print("\nLogin Successful!")
        print("\n===== ATM SERVICES =====")
        print("1. Balance Inquiry")
        print("2. Cash Withdrawal")
        print("3. Exit")

        choice = int(input("Enter your choice (1-3): "))

        if choice == 1:
            print(f"\nAvailable Balance: ₹{balance}")

        elif choice == 2:
            amount = float(input("Enter Withdrawal Amount: ₹"))

            if amount <= 0:
                print("Invalid Amount!")

            elif amount > balance:
                print("Insufficient Balance!")

            else:
                balance -= amount
                print("\nPlease collect your cash.")
                print(f"Withdrawn Amount : ₹{amount}")
                print(f"Remaining Balance: ₹{balance}")

        elif choice == 3:
            print("Thank you for using our ATM.")

        else:
            print("Invalid Choice!")

    else:
        print("Incorrect PIN!")

else:
    print("Invalid Card Number!")

print("\nTransaction Completed.")
```
