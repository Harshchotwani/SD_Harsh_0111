```python
l=[1,2,3,4,4,"sudh","muar",345.345,"abc"]
```

```python
l1_num=[]
l2_num=[]
for i in l:
  if type(i)==int or type(i)==float:
    l1_num.append(i)
  else:
    l2_num.append(i)
print(l1_num)
print(l2_num)
```

```python
number=int(input("enter the number"))
rev=0
temp=number
while number>0:
  digit=number%10
  rev=(rev*10)+digit
  number=number//10
if temp== rev:
  print('the number is palindrome')
else:
  print('the number is not palindrome')
```
